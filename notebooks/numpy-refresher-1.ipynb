## Linear Algebra

```python
a = np.ones((2,3))
b = np.full((3,2), 2)

print(a)
print(b)
```

```python
# Multiplying matrices
c = np.matmul(a,b)
c
```

```python
# Find determinant
d = np.identity(3)
print(np.linalg.det(d))
print(np.linalg.det(c))
```

## Reorganizing Arrays

```python
before = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print(before)
print(before.shape)
```

```python
after1 = before.reshape(4,2)
after1
```

```python
after2 = before.reshape(2, 2, 2)
after2
```

```python
# Vertically Stacking Vectors

v1 = np.array([1, 2, 3, 4])
v2 = np.array([5, 6, 7, 8])

v3 = np.vstack([v1, v2])
v3
```

```python
# Horizontal Stacking

v1 = np.hstack([v1, v2])
v1
```
